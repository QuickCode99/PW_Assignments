# SEABORN Assignment:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.

# Generate synthetic data
np.random.seed(1)
x = np.random.rand(100) * 10  # random values from 0 to 10
y = 3 * x + np.random.randn(100) * 5  # linear relation with noise

# Create DataFrame
df = pd.DataFrame({'X': x, 'Y': y})

# Create scatter plot
sns.set(style="whitegrid")
sns.scatterplot(x='X', y='Y', data=df, color='crimson')

plt.title('Scatter Plot: Synthetic Variables')
plt.xlabel('X Variable')
plt.ylabel('Y Variable')
plt.show()
```

```python
#2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.

# Generate random numbers (normally distributed)
np.random.seed(10)
data = np.random.normal(loc=0, scale=1, size=1000)  # mean = 0, std = 1

# Plot the distribution
sns.set(style="darkgrid")
sns.histplot(data, kde=True, color="steelblue", bins=30)

plt.title("Distribution of a Numerical Variable")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#3. Create a dataset representing categories and their corresponding values. Compare different categories
# based on numerical values

# Create the dataset
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value': [25, 40, 30, 35, 20]
}
df = pd.DataFrame(data)

# Plot the bar chart
sns.set(style="whitegrid")
sns.barplot(x='Category', y='Value', data=df, palette='mako')

plt.title("Category Comparison Based on Values")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

```python
#4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical
# variable across different categories

# Generate synthetic dataset
np.random.seed(42)
categories = np.random.choice(['A', 'B', 'C', 'D'], 200)
values = np.random.normal(loc=50, scale=10, size=200)  # Normal distribution

df = pd.DataFrame({'Category': categories, 'Value': values})

# Box plot
sns.set(style="whitegrid")
sns.boxplot(x='Category', y='Value', data=df, palette='Set2')

plt.title("Distribution of Values Across Categories (Box Plot)")
plt.show()
```

```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using
# a heatmap

# Set seed for reproducibility
np.random.seed(100)

# Generate base variable
x = np.random.normal(50, 10, 200)

# Create correlated variables
data = {
    'Feature_1': x,
    'Feature_2': x + np.random.normal(0, 5, 200),
    'Feature_3': x * 0.5 + np.random.normal(0, 3, 200),
    'Feature_4': np.random.normal(50, 10, 200)  # not correlated
}

df = pd.DataFrame(data)

# Compute correlation matrix
corr_matrix = df.corr()

# Plot heatmap
sns.set(style="white")
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1, linewidths=0.5)

plt.title("Correlation Matrix Heatmap")
plt.show()
```

# PLOTLY ASSIGNMENT:

```python
#1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a
# threedimensional space

import numpy as np
import pandas as pd
import plotly.graph_objs as go

# Set seed and generate synthetic 3D data
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

# Create 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=df['X'],
    y=df['Y'],
    z=df['Z'],
    mode='markers',
    marker=dict(
        size=4,
        color=df['Z'],  # color by Z value
        colorscale='Viridis',
        opacity=0.8
    )
)])

# Layout
fig.update_layout(
    title='3D Scatter Plot of Synthetic Data',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)

fig.show()
```

```python
#2.  Using the Student Grades, create a violin plot to display the distribution of scores across different
# grade categories

import numpy as np
import pandas as pd
import plotly.express as px

# Generate student grade data
np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df_grades = pd.DataFrame(data)

# Create violin plot
fig = px.violin(df_grades, x='Grade', y='Score', box=True, points='all',
                title='Distribution of Student Scores by Grade',
                color='Grade', color_discrete_sequence=px.colors.qualitative.Set3)
fig.show()
```

```python
# Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.

# Generate sales data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_sales = pd.DataFrame(data)

# Create a pivot table for heatmap
pivot = df_sales.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Create heatmap using Plotly
fig = px.imshow(pivot,
                labels=dict(x="Month", y="Day", color="Avg Sales"),
                x=pivot.columns,
                y=pivot.index,
                color_continuous_scale='Viridis',
                title='Sales Heatmap by Day and Month')

fig.update_layout(yaxis_nticks=30)
fig.show()
```

```python
#3.  Using the sales data, generate a heatmap to visualize the variation in sales across different months and
# days.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(20)

# Generate random sales data
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df = pd.DataFrame(data)

# Pivot the data: rows = Day, columns = Month, values = average Sales
heatmap_data = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.0f', linewidths=0.5)

# Add labels and title
plt.title("Heatmap of Sales by Day and Month")
plt.xlabel("Month")
plt.ylabel("Day")

plt.tight_layout()
plt.show()
```

# BOKEH ASSIGNMENT:

```python
#1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

from bokeh.plotting import figure, show, output_notebook
import numpy as np

# Display Bokeh plots inside a notebook
output_notebook()

# Generate data
x = np.linspace(0, 10, 200)
y = np.sin(x)

# Create Bokeh figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=700, height=400)
p.line(x, y, line_width=2, color='navy', legend_label='sin(x)')

# Show the plot
show(p)
```

```python
#2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
# markers based on the 'sizes' and 'colors' columns

from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
import numpy as np

output_notebook()

# Generate random data
np.random.seed(42)
N = 100
x = np.random.rand(N)
y = np.random.rand(N)
sizes = np.random.randint(10, 30, size=N)
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=N)

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y, size=sizes, color=colors))

# Create the scatter plot
p = figure(title="Bokeh Scatter Plot with Sizes and Colors", width=700, height=400)
p.scatter(x='x', y='y', size='size', color='color', alpha=0.7, source=source)

show(p)
```

```python
#3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

from bokeh.plotting import figure, show, output_notebook
from bokeh.transform import dodge
from bokeh.models import ColumnDataSource
import pandas as pd

output_notebook()

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Convert to DataFrame and ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create figure
p = figure(x_range=fruits, height=400, title="Fruit Counts", toolbar_location=None, tools="")

# Bar chart
p.vbar(x='fruits', top='counts', width=0.6, source=source, color="tomato")

# Axis and grid styling
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = "Count"

show(p)
```

```python
#4. Create a Bokeh histogram to visualize the distribution of the given data.

from bokeh.plotting import figure, show, output_notebook
import numpy as np

output_notebook()

# Generate data
data_hist = np.random.randn(1000)  # Normally distributed
hist, edges = np.histogram(data_hist, bins=30)

# Create the histogram plot
p = figure(title="Histogram of Normally Distributed Data", width=700, height=400)
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="skyblue", line_color="white")

# Axis labels
p.xaxis.axis_label = 'Value'
p.yaxis.axis_label = 'Frequency'

show(p)
```

```python
#5. Create a Bokeh heatmap using the provided dataset.

from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, LinearColorMapper
from bokeh.transform import linear_cmap
import numpy as np

output_notebook()

# Generate data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten data for plotting
source = ColumnDataSource(data=dict(
    x=xx.flatten(),
    y=yy.flatten(),
    value=data_heatmap.flatten()
))

# Color mapper
mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

# Create figure
p = figure(title="Bokeh Heatmap", x_range=(0, 1), y_range=(0, 1),
           width=500, height=500, tools="", toolbar_location=None)

# Draw heatmap rectangles
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
       fill_color=linear_cmap('value', 'Viridis256', data_heatmap.min(), data_heatmap.max()),
       line_color=None)

# Axes and grid
p.xaxis.axis_label = "X"
p.yaxis.axis_label = "Y"
p.grid.grid_line_color = None

show(p)
```
